## Visualizations

### 1) The season might affect the number of the shared tweets. This can either be confirmed or denied by analysing the number of tweets in different seasons depending on month column.

```python
num_post= tweets_only['month'].value_counts().plot(kind= 'bar', color= '#14C136', title= 'The Number of Posted Tweets in each Month');
num_post.set_xlabel('Month');
num_post.set_ylabel('Number of Tweets');               #Creating a bar chart to show the number of the posted tweets in different seasons depending on month
```

### It is clear that the highest number of posted tweets is in December (Winter season).

### 2) The number of the retweet_count might be related to the rating_numerator of the dog.

```python
rate_retweet= df.groupby('rating_numerator')['retweet_count'].sum().plot(kind= 'bar', color= '#9820F0', title= 'The Number of Retweets and the Rating of the Dog');
rate_retweet.set_xlabel('Dog Rating');
rate_retweet.set_ylabel('Number of Retweets');       #Creating a bar chart to show the dogs' ratings and the number of retweeded posts that have these dogs' ratings
```

### It is clear that the rating of the dog does not have a strong relation with the number of its retweeted posts. As the ratings of 14 and above have lower number of retweets than the ratings of 13 and below.
